# **SETS**

BASIC SET CREATION

```python
#Create a set of 5 colors and print it
x = {"black" , "white" , "pink" , "orange" , "blue"}
print(x)
```

```python
#Create a set with duplicates and check behavior
y = {1,2,3,1,4,5,4}
print(y)
```

```python
#Loop through a set and print each element
z= {"apple" , "banana" , "cherry"}
for i in z:
  print(i)
```

ADDING AND REMOVING ELEMENTS

```python
#Create a set and use .add()
animals = {"lion", "tiger", "elephant"}
animals.add("giraffe")
print(animals)
```

```python
#Remove element using .remove()
nums = {1,2,3,5}
nums.remove(3)
print(nums)
```

```python
#Remove using .discard() and compare
num = {10, 30 , 60, 90}
nums.discard(20)
print(nums)
nums.discard(50)
print(nums)
```

```python
#Remove a random element using .pop()
alpha = {"a", "b", "c", "d", "e"}
removed = alpha.pop()
print("REMOVED :" , removed)
print("REMANING :" , alpha)
```

SET OPERATIONS

```python
#Create two sets and find their union
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
union = a.union(b)
print(union)
```

```python
#Create two sets and find their intersection
a = {10, 20, 30, 40}
b = {30, 40, 50, 60}
intersection = a.intersection(b)
print(intersection)
```

```python
#Find difference (A - B)
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
c= a.difference(b)
print(c)
```

OTHER METHODS

```python
#Use .copy() and modify copy
a = {1, 2, 3, 4}
b = a.copy()
b.add(5)
print("ORIGINAL: ", a)
print("COPY: ", b)
```

```python
#Use .clear() to empty a set
s = {1, 2, 3, 4, 5}
s.clear()
print(s)
```

```python
#Convert list with duplicates into set
l= [1,2,3,4,3,5,2,3]
s= set(l)
print(s)
```

APPLICATONS

```python
#Find all unique characters in a string
text = input("Enter a string: ")
unique = set(text)
print("Unique characters in the string:", unique )
```

```python
#Find students who play both sports
football = {"ariv", "yash", "om", "shashwat"}
cricket = {"shashwat", "virat" , "rohit","yash"}
both = football.intersection(cricket)
print(both)
```

```python
#Find students who play only cricket
football = {"ariv", "yash", "om", "shashwat"}
cricket = {"shashwat", "virat" , "rohit","yash"}
only = cricket - football
print(only)
```

```python
#Print all unique words in a sentence
sentence = input("Enter a sentence: ")
words = sentence.split()
unique = set(words)
print("Unique words in the sentence:")
for word in unique:
    print(word)
```

```python
#Check if two sets are equal
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
if a == b:
  print("SETS ARE EQUAL")
else:
  print("SETS ARE NOT EQUAL")
```

```python
#Check if two sets have at least one common element
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
common = a.intersection(b)
if common:
  print("SETS HAVE COMMON ELEMENT :", common)
else:
  print("SETS DO NOT HAVE COMMON ELEMENT")
```

```python
#Find number of unique elements in a list
num = [1, 2, 3, 4, 3, 5, 2, 3]
unique = set(num)
print("Number of unique elements:", len(unique))
```
