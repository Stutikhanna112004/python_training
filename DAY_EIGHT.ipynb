```python
#Write a program to find your age by taking your date of birth as input (use datetime.date).
import datetime
today = datetime.date.today()
birth = input("enter your birthday:")
birth_date = datetime.datetime.strptime(birth, "%d/%m/%Y")
age = today.year - birth_date.year
print(age)
```

```python
# Write a Python program to find the number of days left in the current year.
import datetime
today = datetime.date.today()
days = datetime.date(today.year, 12, 31) - today
print(days.days)
```

```python
#Q7. Write a program to add 10 days to the current date using timedelta.
import datetime
today = datetime.date.today()
print(today + datetime.timedelta(10))
```

```python
#Q8. Write a program to find the difference in days between two given dates.
import datetime
a= input("enter a date (DD MM YYYY):")
b= input("enter a date (DD MM YYYY):")
a_date = datetime.datetime.strptime(a, "%d %m %Y")
b_date = datetime.datetime.strptime(b, "%d %m %Y")
c= a_date - b_date
print(c.days)
```

```python
#. Write a Python program to check if today is your birthday or not.
import datetime
today = datetime.date.today
birthday = input("enter your birthday (DD MM YYYY):")
birthday_date = datetime.datetime.strptime(birthday, "%d %m %Y")
if birthday_date == today:
    print("Happy birthday!")
else:
    print("Sorry, today is not your birthday.")
```
