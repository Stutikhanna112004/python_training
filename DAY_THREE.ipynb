```python
#DICTIONARY
diction = { "INDIA" : "NEW DELHI" , "CHINA" : "BEJING" , "FRANCE" : "PARIS" , "JAPAN" : "TOKYO" }
print(diction)
print(diction["INDIA"])      #PRINTING THE VALUE OF A SPECIFIC KEY
diction.update({"GERMANY" : "BERLIN"})   #ADDING NEW KEY & VALUE PAIR
print(diction)
diction.pop("FRANCE")    #REMOVING A KEY AND VALUE PAIR
print(diction)
del diction["CHINA"]    #DELETING A KEY AND ITS VALUE
print(diction)
diction.clear()         #CLEARING A DICTIONARY
print(diction)
```

```python
data= { "RAM" : 87 , "SHYAM" : 88 , "KRISH " : 90 }
print(data.keys())          #RETURNS ALL THE KEYS ONLY
print(data.values())        #RETURNS ALL THE VALUES ONLY
print(data.items())         #RETURNS ALL THE KEY & VALUE PAIR
print(data.get("RAHUL"))     #RETURNS THE VALUE IF IT EXISTS AND RETURNS NONE IT DOESN'T AVOIDING ERROR
data.update({"RAHUL" : 99, "YASH" : 90, "OM" : 77})
print(data)
```

```python
#WAP TO MAKE COPY OF A DICTIONARY AND UPDATE THE COPIED DICTIONARY WITHOUT MAKING ANY CHANGES IN ORIGNAL DICTIONARY
data= { "RAM" : 87 , "SHYAM" : 88 , "KRISH " : 90 }
print(data)
copy = data.copy()
copy.update({"RAHUL" : 99, "YASH" : 90, "OM" : 77})
print(copy)
```

```python
#WAP TO COUNT THE FREQUENCY OF WORDS IN A SENTENCE IN A DICTIONARY
sentence = "Python is a scary snake but Python language is easy"
words = sentence.split()
frequency = {}
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
print("word frequency :" , frequency)
for word,count in frequency.items():
    print(word , ":", count)
```

```python
#TO FIND THE MAXIMUM NUMBER
def max(a,b):
    if a>b:
        print("max :" , a)
    else:
        print("max :" , b)
x= int(input("ENTER A NUMBER : "))
y= int(input("ENTER A NUMBER : "))
max(x,y)
```

```python
#TO FIND AVERAGE OF 4 NUMBERS
def avg(a,b,c,d):
  avg=(a+b+c+d)/4
  print("avg :" , avg)
avg(2,3,4,5)
```

```python
#FUNCTION TO COUNT ALL THE VOWELS IN A STRING
def countvowels(w):
  count = 0
  vowels = "aeiouAEIOU"
  for char in w:
    if char in vowels:
      count += 1
  print("vowels :" , count)
x= input("ENTER A WORD : ")
countvowels(x)
```

```python
#TO FIND FACTORIAL A NUMBER USING LOOPS
def fact(n):
  t=1
  for i in range(1,n+1):
    t*=i
  print("fact :" , t)
x= int(input("ENTER A NUMBER : "))
fact(x)
```

```python
#TO CHECK IF A NUMBER IS A PANINDROME OR NOT
def palindrome(word):
    if word == word[::-1]:
        return True
    else:
        return False
w= input(print("ENTER A STRING:"))
palindrome(w)
```

```python
#RECURSION
#WAP A RECURSIVE FUNCTION TO FIND FACTORIAL OF A NUMBER
def facto(n):
  if(n==0 or n==1):
    return 1
  else:
    return n*facto(n-1)
x= int(input("ENTER A NUMBER : "))
fact(x)
```

```python
#WAP A RECURSIVE FUNCTION TO PRINT FIBONACCI SERIES UP TO N TERMS
def fib_recursive(n):
    if n <= 1:
        return n
    else:
        return (fib_recursive(n-1) + fib_recursive(n-2))
def print_fibonacci_series(num_terms):
    if num_terms <= 0:
        print("Please enter a positive integer.")
    else:
        print("Fibonacci Series:")
        for i in range(num_terms):
            print(fib_recursive(i), end=" ")

nterms = int(input("Enter the number of terms: "))
print_fibonacci_series(nterms)
```
