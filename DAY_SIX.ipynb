# **POLYMORPHISM**


```python
class animal:
  def sound(self):
    print("some generic sound")
class cat(animal):
  def sound(self):
    print("meow")
a= animal()
c=cat()
a.sound()
c.sound()
```

```python
class shape:
    def draw(self):
      print(f"drawing shape")
class circle(shape):
  def draw(self):
    print("drawing circle")
class square(shape):
  def draw(self):
    print("drawing square")
a=shape()
c=circle()
s=square()
a.draw()
c.draw()
s.draw()
```

```python
class animal:
  def sound(self):
    print("some generic sound")
class cow(animal):
  def sound(self):
    print("moo")
class lion(animal):
  def sound(self):
    print("roar")
a=animal()
c=cow()
l=lion()
a.sound()
c.sound()
l.sound()
```

```python
class printer:
  def use(self,doc):
    print(f"printing {doc}")
class scanner:
  def use(self,doc):
    print(f"scanning {doc}")
p=printer()
s=scanner()
p.use("book")
s.use("book")
```

```python
class vehicle:
  def move(self):
    print("vehicle started")
class bike(vehicle):
  def move(self):
    print("bike started")
class car(vehicle):
  def move(self):
    print("car started")
v=vehicle()
b=bike()
c=car()
v.move()
b.move()
c.move()
```

```python
class principal:
  def role(self):
    print("role is principal")
class teacher:
  def role(self):
    print("role is teacher")
class student:
  def role(self):
    print("role is student")
p=principal()
t=teacher()
s=student()
school=[p,t,s]
for obj in school:
  obj.role()
```

```python
class walk:
  def showaction(self):
    print("you are walking")
class run:
  def showaction(self):
    print("you are running")
class jump:
  def showaction(self):
    print("you are jumping")
w=walk()
r=run()
j=jump()
w.showaction()
r.showaction()
j.showaction()
```

# **ENCAPSULATION**
 binding data and methods that operate on that data into a single unit (class) and restricting direct access to some components

```python
class bank:
  def __init__(self,owner,balance):
    self.owner=owner
    self.__balance=balance
  def deposit(self,amount):
    self.__balance+=amount
  def balance(self):
    return self.__balance
  def withdraw(self,amount):
    if amount<=self.__balance:
      self.__balance-=amount
acc = bank("ram",1000)
acc.deposit(500)
acc.withdraw(200)
print(acc.balance())
```

```python
class student:
  def __init__(self,name,marks):
    self.name=name
    self.__marks=marks
  def display(self):
    print("name:",self.name)
    print("marks:",self.__marks)
s=student("ram",90)
s.display()
```

```python
class employee:
  def __init__(self,name,salary):
    self.name=name
    self.__salary=salary
  def salary(self, bonus):
     self.__salary += bonus
  def displaysalary(self):
    return self.__salary
e=employee("ram",50000)
e.salary(10000)
print(e.displaysalary())
```

```python
class mobile:
  def __init__(self,name,price):
    self.name=name
    self.__price=price
  def chageprice(self,discount):
    self.__price-=(discount/100)*self.__price
  def getprice(self):
    return self.__price
m=mobile("samsung",100000)
m.chageprice(10)
print(m.getprice())
```

```python
class person:
  def __init__(self,name,age):
    self.name=name
    self.__age=age
  def showage(self):
    return self.__age
p=person("ram",20)
print(p.showage())
```

```python
class laptop:
    def __init__(self,ram):
        self.__ram=ram
    def showram(self):
        return self.__ram
r=laptop(32)
print(r.showram())
```

```python
class car:
  def __init__(self,name,speed):
    self.name=name
    self.__speed= speed
  def setspeed(self,incspeed):
    self.__speed+=incspeed
  def showspeed(self):
    return self.__speed
c=car("bmw",100)
c.setspeed(10)
print(c.showspeed())
```

```python
class acc:
  def __init__(self,name,pin):
    self.name=name
    self.__pin=pin
  def checkpin(self,enpin):
    if(self.__pin==enpin):
      return True
    else:
      return False
  def __str__(self):
    return self.name

a=acc("ram",1234)
print(a.__str__())
print(a.checkpin(1235))
```

```python
class lib:
  def __init__(self,totbooks):
    self.__totbooks = totbooks
  def display(self):
    print(f"Total books in library are {self.__totbooks}")
l=lib(10000)
l.display()
```

```python
class product:
  def __init__(self,name,price):
    self.name = name
    self.__price = price
  def upprice(self,upprice):
    self.__price += upprice
  def getprice(self):
    return self.__price
p=product("laptop", 50000)
p.upprice(10000)
print(p.getprice())
```
