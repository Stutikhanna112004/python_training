# Smart Number Guessing Game

IMPORTING MODULES

```python
import random     #THIS MODULE WILL GENERATE A RANDOM NO.
import os         #CHECKS IF A FILE EXISTS IN THE COMPUTER
import time      #THIS MODULE WILL CALCULATE THE TIME USER TOOK
scorefile = "highscore.txt"     #STORES THE FILE NAME WHERE HIGH SCORES WILL BE SAVED
```

```python
def loadhighscore():
  if os.path.exists(scorefile):   #CHECKING IF FILE EXISTS OR NOT
    with open(scorefile, "r") as f:    #OPENING THE FILE IN  READ MODE
      return int(f.read())
  else:
    return 0
```

```python
def savehighscore(score):
  with open(scorefile, "w") as f:     #OPENING FILE IN WRITE MODE
    f.write(str(score))           #WRITING THE SCORE IN FILE AS TXT
```

```python
def playgame():
  print("\nWelcome to the Smart Number Guessing Game!")
  print("\nChoose difficulty:")
  print("1. Easy(1-20, 10 attempts)")
  print("2. Medium(1-50, 7 attempts)")
  print("3. Hard(1-100, 5 attempts)")
  choice = input("Enter your choice (1/2/3): ")
  if choice == "1":
    maxnum = 20
    attemptsallowed = 10
  elif choice == "2":
    maxnum = 50
    attemptsallowed = 7
  elif choice == "3":
    maxnum = 100
    attemptsallowed = 5
  else:
    print("Invalid choice. Please try again.")
    return
  number = random.randint(1, maxnum)   #SELECTS RANDOM NUM B/W 1 AND MAXNUM
  attempts = 0           #COUNTING FOR NUM OF GUESSES
  print(f"\nI'm thinking of a number between 1 and {maxnum}.")
  print(f"\nYou have {attemptsallowed} attempts to guess the number.")
  starttime = time.time()           #STORES CURRENT TIME
  while attempts < attemptsallowed:       #LOOP UNTILL ATTEMOTS ARE USED
    try:
      guess = int(input("Enter your guess: "))
    except ValueError:
      print("Invalid input. Please enter a number.")  #PREVENTS ERROR
    attempts += 1
    if guess < number:
      print("Too low!")
    elif guess > number:
      print("Too high!")
    elif guess == number:
      endtime = time.time()
      timetaken = round(endtime - starttime,2)    #CALCULATES TOTAL TIME SPENT
      score = max(0, 1000 - (attempts - 100 )- int(timetaken * 10))  #SCORE DECREASES WITH MORE ATTEMPTS AND TIME
      print(f"\nCongratulations! You guessed the number in {attempts} attempts.")
      print(f"Time taken: {timetaken} seconds")
      print(f"Score: {score}")
      highscore = loadhighscore()          #READS PREV HIGHSCORE
      if score > highscore:               #SAVES NEW HIGHSCORE
        savehighscore(score)
      else:
        print(f"High score: {highscore}")
      break
  else:
        print("Game Over!")
        print(f"The number was {number}.")
while True:
        playgame()
        again = input("\n Play again? (y/n): ")
        if again.lower() != "y":
            print("Thanks for playing!")
            break
```
