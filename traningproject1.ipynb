# **Student Result Management** **System**

Install & Download Datase

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("spscientist/students-performance-in-exams")

print("Path to dataset files:", path)
```

import libraries

```python
import pandas as pd
import json
import csv
import os
```

```python
DATA_JSON = "/content/students.json"
```

```python
CSV_SOURCE = "/content/students_report_manual.csv"
```

grade calculator

```python
def calculate_grade(p):
    if p >= 90:
        return "A+"
    elif p >= 80:
        return "A"
    elif p >= 70:
        return "B"
    elif p >= 60:
        return "C"
    elif p >= 50:
        return "D"
    else:
        return "Fail"
```

initalize dataset

```python
def initialize_dataset():
    df = pd.read_csv(CSV_SOURCE)
    students = {}

    for _, row in df.iterrows():
        roll = str(row["ID"])
        name = row["Name"]

        marks = {
            "English": int(row["English"]),
            "Math": int(row["Math"]),
            "Physics": int(row["Physics"]),
            "Science": int(row["Science"])
        }

        total = sum(marks.values())
        subjects = len(marks)
        percentage = total / subjects
        grade = calculate_grade(percentage)

        students[roll] = {
            "name": name,
            "marks": marks,
            "percentage": round(percentage, 2),
            "grade": grade
        }

    with open(DATA_JSON, "w") as f:
        json.dump(students, f, indent=4)

    print("Dataset initialized successfully!")
```

load data

```python
def load_data():
    if os.path.exists(DATA_JSON):
        with open(DATA_JSON, "r") as file:
            return json.load(file)
    return {}
```

save data

```python
def save_data(data):
    with open(DATA_JSON, "w") as file:
        json.dump(data, file, indent=4)
```

add student

```python
def add_student(data):
    roll = input("Enter Roll Number: ")

    if roll in data:
        print("Student already exists!")
        return

    name = input("Enter Name: ")

    marks = {}
    subjects = ["English", "Math", "Physics", "Science"]

    for sub in subjects:
        marks[sub] = int(input(f"Enter marks for {sub}: "))

    percentage = sum(marks.values()) / len(marks)
    grade = calculate_grade(percentage)

    data[roll] = {
        "name": name,
        "marks": marks,
        "percentage": round(percentage, 2),
        "grade": grade
    }

    save_data(data)
    print("âœ… Student added successfully!")
```

Update Marks

```python
def update_marks(data):
    roll = input("Enter Roll Number: ")

    if roll not in data:
        print("Student not found!")
        return

    marks = data[roll]["marks"]

    for subject in marks:
        new_mark = int(input(f"Enter new marks for {subject}: "))
        marks[subject] = new_mark

    percentage = sum(marks.values()) / len(marks)
    grade = calculate_grade(percentage)

    data[roll]["percentage"] = round(percentage, 2)
    data[roll]["grade"] = grade

    save_data(data)
    print("Marks updated successfully!")
```

Search Student

```python
def search_student(data):
    roll = input("Enter Roll Number: ")

    if roll not in data:
        print("Student not found!")
        return

    s = data[roll]

    print("\n------ Student Record ------")
    print("Name:", s["name"])
    print("Marks:")
    for sub, mark in s["marks"].items():
        print(f"   {sub}: {mark}")
    print("Percentage:", s["percentage"])
    print("Grade:", s["grade"])
```

 Export to CSV

```python
def export_to_csv(data):
    csv_path = "/content/results.csv"

    with open(csv_path, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Roll", "Name", "English", "Math", "Physics", "Science", "Percentage", "Grade"])

        for roll, info in data.items():
            marks = info["marks"]
            writer.writerow([
                roll,
                info["name"],
                marks["English"],
                marks["Math"],
                marks["Physics"],
                marks["Science"],
                info["percentage"],
                info["grade"]
            ])

    print("Exported to:", csv_path)
```

Main Program

```python
def main():
    if not os.path.exists(DATA_JSON):
        initialize_dataset()

    data = load_data()

    while True:
        print("\nStudent Result Management ")
        print("1. Add Student")
        print("2. Update Marks")
        print("3. Search Student")
        print("4. Export CSV")
        print("5. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_student(data)
        elif choice == "2":
            update_marks(data)
        elif choice == "3":
            search_student(data)
        elif choice == "4":
            export_to_csv(data)
        elif choice == "5":
            print("Program Closed")
            break
        else:
            print("Invalid choice!")


main()
```
