```python
#LIST
#CREATING THE LIST AND PRINTING THE FIRST, LAST AND THE LENGTH OF THE LIST
x = [1,2,3,4,5]
print ("first element:", x[0])
print ("last element: ", x[-1])
print ("length :", len(x))
```

```python
#sum and avg of the list
x= [1,2,3,4,5,6]
tot = sum(x)
avg = tot/len(x)
print("sum:", tot)
print("avg:,", avg)
```

```python
#insert and delete an element
x=["apple","kiwi", "orange"]
x.append("melon")
print(x)
x.remove("apple")
print(x)
x.pop()
print (x)
```

```python
#finding count of duplicates of an element in a list
x=[1,2,3,4,5,3,6,3,3]
print("duplicate count:" ,x.count(3))
```

```python
#giving the index of an element
x=[1,2,3,4,5]
y=int(input("enter a number:"))
print("index of",y,"is",x.index(y))
```

```python
#sorting in ascending and descending
x=[4,1,2,6,5,8,3,9]
print("sorted list: " ,sorted(x))
print("reverse list: " ,sorted(x, reverse=True))
```

```python
#converting a list to tuple
x=[1,2,3,4,5]
print("list :",x)
print("LIST TO TUPLE", tuple(x))
```

```python
#creating a tuple and converting it into list and modifing it and convert it back to a tuple
x=(1,2,3,4,5)
y=list(x)
print("TUPLE TO LIST", y)
y.append(6)
print("LIST TO TUPLE", tuple(y))
```

```python
#creating a nested tuple and printing the elements inside it
tuple1=((1,2,3),("stuti","ram","shyam"),(6,7,8,9))
for i in range(len(tuple1)):
    for j in range(len(tuple1[i])):
      print(tuple1[i][j])
```
