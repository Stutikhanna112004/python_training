```python
#Count total number of words in the file.
import string
with open("/stutidata.txt","r") as f:
    file_contents = f.read()
for ch in string.punctuation:
    file_contents = file_contents.replace(ch, "")
whilespaces = file_contents.split()
print(len(whilespaces))
```

```python
#Append 2 more lines into the existing file without deleting previous content.
with open("/stutidata.txt","a") as f:
  f.write("\nStuti , 19 , Lucknow\nSusmita , 19 , Basti")
with open("/stutidata.txt","r") as f:
    file_contents = f.read()
    print(file_contents)
```

```python
#Count number of characters (excluding spaces)
with open("/stutidata.txt","r") as f:
    file_contents = f.read()
    print(len(file_contents.replace(" ","")))
```

```python
#Open the file in r+ mode and update the first studentâ€™s name from "Amit" to "Amit Kumar".
with open("/stutidata.txt","r+") as f:
    initial_contents = f.read()
    updated_contents = initial_contents.replace("Amit","Amit Kumar")
    f.seek(0)
    f.truncate()
    f.write(updated_contents) # Write the modified content
    f.seek(0) # Go to the beginning again to read for printing
    final_file_contents = f.read()
    print(final_file_contents)
```
