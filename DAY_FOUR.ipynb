# **OPPs**

```python
#OOPs
class student:
  def __init__(self,name,age):
    self.name=name
    self.age=age
s1 = student("ram",20)
s2 = student("shyam",21)
print(s1.name,s1.age)
print(s2.name,s2.age)
```

```python
class car:
  def __init__(self,brand,price):
    self.brand=brand
    self.price=price
  def display(self):
    print(self.brand,self.price)
c1 = car("audi",100000)
c2 = car("bmw",200000)
c1.display()
c2.display()
```

```python
class employee:
  def __init__(self,name,salary):
    self.name=name
    self.salary=salary
  def display(self):
    print(self.name,self.salary)
e1=employee("ram",10000)
e2=employee("shyam",20000)
e1.display()
e2.display()
```

```python
class mobile:
  def __init__(self,brand,model):
    self.brand=brand
    self.model=model
  def display(self):
    print(self.brand, self.model)
m1 = mobile("samsung","s22")
m2 = mobile("apple","iphone14")
m3 = mobile("motorola","motog60")
m1.display()
m2.display()
m3.display()
```

```python
class book:
  def __init__(self,title,author):
    self.title = title
    self.author = author
  def display(self):
    print("book:",self.title,",author:",self.author)
b1=book("wings of fire","apj abdul kalam")
b2=book("the alchemist","paulo coelho")
b1.display()
b2.display()
```

# **INHERITANCE**

```python

class person:
    def __init__(self,fname,lname,age):
        self.fname=fname
        self.lname=lname
        self.age = age
class student(person):
  def __init__(self,fname,lname,age,rollno,section):
    person.__init__(self,fname,lname,age)
    self.rollno=rollno
    self.section=section
  def show(self):
       print("person details:", "first name -", self.fname,"last name -", self.lname, "age -" , self.age, "rollno-", self.rollno, "section-", self.section)
s1=student("ram","sharma",20,101,"a")
s2=student("shyam","sharma",21,102,"b")
s1.show()
s2.show()
```

```python
class shape:
    def draw(self):
        print("there are different shapes")
class circle(shape):
    def drawing(self):
        print("circle is drawing")
c=circle()
c.draw()
c.drawing()
```

```python
class teacher:
  def teach(self):
    print("teacher is teaching")
class mathteacher(teacher):
  def study(self):
    print("math teacher is teaching maths")
m=mathteacher()
m.teach()
m.study()
```

```python
class computer:
  def power_on(self):
    print("computer is powering on")
class laptop(computer):
  def display(self):
    print("laptop is powering on")
l=laptop()
l.power_on()
```

```python
class mobile:
  def call(self):
    print("mobile is calling")
class Smartphone(mobile):
  def display(self):
    print("smartphone is calling")
s=Smartphone()
s.call()
s.display()
```
